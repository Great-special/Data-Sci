### Problem Statement 2: Write a Python program to perform the following tasks

1. Create a pandas series from the below dictionary where indices are subjects

```python
subjects_std_score = {
'English':{'Sam':60,'Jackson':74,'Ahree':85},
'History':{'Gloria':83,'Sam':65,'Isla':78,'Aron':72,'Gray':61},
'Geography':{'Jackson':92,'Gloria':95,'Isla':82,'Aron':75,'Ahree':76},
'Mathematics':{'Sam':99,'Gloria':74,'Jackson':89,'Ahree':85,'Gray':95},
'Science':{'Sam':89,'Aron':82,'Gray':78,'Isla':93,'Ahree':87}
}
```

```python
subject_series = pd.Series(subjects_std_score)
subject_series
```

2. Convert the created series into DataFrame and replace the null values with zeroes.

```python
df = subject_series.apply(pd.Series)
df
```

```python
df = df.fillna(0)
df
```

3. Transpose the DataFrame and create a new column 'Average' and fill the values in it by calculating the average of all subjects.

```python
df1 = df.transpose()
df1
```

```python
df1.info()
```

```python
df1['Average'] = df1.mean(axis=1)
df1
```

### Problem Statement 3: Write a Python program to create a series from 1 to 1000 and select only numbers divisible by 7 and 17.

```python
num_range = np.arange(1,1001)
range_series = pd.Series(num_range)
range_series
```

```python
filtered_numbers = range_series[(range_series % 7 == 0) & (range_series % 17 == 0)]
```

### Problem Statement 4: Sylphia has a dataset of various cereals sold in the supermarket.

#### Tasks to be performed: 
1. Import the dataset. 

```python
cereal_df = pd.read_csv("cereal.csv")
cereal_df
```

2. Plot ratings of different types of manufacturers. 

```python
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
cereal_df.columns
```

```python
manuf_rating = cereal_df.groupby(['mfr'])[['rating']].sum().reset_index()
manuf_rating
```

```python
plt.figure(figsize=(10, 6))
plt.barh(manuf_rating['mfr'], manuf_rating['rating'], color='skyblue')
plt.ylabel('Manufacturer')
plt.xlabel('Rating')
plt.title('Ratings of Different Manufacturers')
plt.show()
```

3. Use xticks range form 0-100.

```python
manuf_rating_count = cereal_df.groupby(['mfr'])[['rating']].count().reset_index()
manuf_rating_count
```

```python
plt.figure(figsize=(10, 6))
plt.barh(manuf_rating_count['mfr'], manuf_rating_count['rating'], color='skyblue')
plt.ylabel('Manufacturer')
plt.xlabel('Rating')
plt.title('Ratings of Different Manufacturers')
plt.xticks(range(0, 100, 10))
plt.show()
```

4. Change the style of the graph to seaborn

```python
# Set Seaborn style
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
plt.barh(manuf_rating_count['mfr'], manuf_rating_count['rating'], color='skyblue')
plt.ylabel('Manufacturer')
plt.xlabel('Rating')
plt.title('Ratings of Different Manufacturers')
plt.xticks(range(0, 100, 10))
plt.show()
```
