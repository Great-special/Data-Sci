## Mini Project 1- Stock Market Analysis Using Python

##### Objective: Use Python libraries such as Pandas, Seaborn, and Matplotlib to extract and analyze the information, visualize it, and look at different ways to analyze the risk of a stock, based on its performance history.

##### Import the necessary libraries:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import yfinance as yf
```

##### Read the data from Yahoo Finance:

```python
# Define the stock symbols and time period
symbols = ['AAPL', 'GOOGL', 'MSFT', 'AMZN']
start_date = '2020-01-01'
end_date = '2023-12-31'

# Fetch stock data from Yahoo Finance
stocks_data = yf.download(symbols, start=start_date, end=end_date)
stocks_data
```

##### 2. Perform cleaning:

```python
# Drop missing values
stocks_data.dropna(inplace=True)
```

##### 3. Calculate the change in stock price over time:

```python
# Calculate daily returns
daily_returns = stocks_data['Adj Close'].pct_change()
daily_returns
```

##### 4. Visualize the change in a stock’s volume over time:

```python
# Visualize the change in a stock’s volume being traded over time
plt.figure(figsize=(10, 6))

for symbol in symbols:
    sns.lineplot(data=stocks_data['Volume'][symbol], label=symbol, linewidth=1.2)

plt.title('Stock Volume Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Volume', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

##### 5. Calculate the moving average of various stocks:

```python
# Calculate moving averages
moving_averages = stocks_data['Adj Close'].rolling(window=30).mean()
moving_averages
```

##### 6. Calculate the daily return average of a stock:

```python
# Calculate average daily returns
avg_daily_returns = daily_returns.mean()
avg_daily_returns
```

##### 7. Add a new column 'Trend' based on the 'Daily Return':

```python
stocks_data.columns
```

```python
print(stocks_data.shape)
```

```python
# Utilize np.select for efficient condition-based assignment
one_d_array = daily_returns.values.flatten()

conditions = [
    (one_d_array > 0),
    (one_d_array < 0),
    (one_d_array == 0)
]

# Specify corresponding choices for each condition
choices = ['Up', 'Down', 'No Change']

# Employ np.select to efficiently apply conditions and assign values to 'Trend' column
flatten_trend = np.select(conditions, choices, default='No Change')

# Assign the reshaped 'Trend' values back to the DataFrame
stocks_data[('Trend', 'AAPL')] = flatten_trend[:1006]
stocks_data[('Trend', 'AMZN')] = flatten_trend[1006:2*1006]
stocks_data[('Trend', 'GOOGL')] = flatten_trend[2*1006:3*1006]
stocks_data[('Trend', 'MSFT')] = flatten_trend[3*1006:]
stocks_data.head()
```

##### 8. Visualize trend frequency through a Pie Chart:

```python
tickers = ['AAPL', 'AMZN', 'GOOGL', 'MSFT']

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

# Plot pie charts for each stock
for i, ticker in enumerate(tickers):
    trend_counts = stocks_data[('Trend', ticker)].value_counts()
    axes[i // 2, i % 2].pie(trend_counts, labels=trend_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
    axes[i // 2, i % 2].set_title(f'Trend Frequency - {ticker}')

plt.show()
```

##### 9. Calculate the correlation between daily returns of different stocks:

```python
# Calculate correlation matrix
correlation_matrix = daily_returns.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10,6))   
sns.heatmap(correlation_matrix, center=0, cmap='coolwarm', annot=True)
plt.show()
```
